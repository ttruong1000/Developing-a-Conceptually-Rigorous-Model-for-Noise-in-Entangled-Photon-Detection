```python
# Import necessary Python packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define necessary parameters (as constants)
P_D = 1e-6  # dark count probability (real number between 0 and 1)
ETA = 0.1   # detector efficiency (real number between 0 and 1)
MU  = 0.02  # mean number of photons detected on some time interval (positive real number)
```

```python
# Fidelities based on varying dark count probability
fidelities_PD1 = []
fidelities_PD2 = []
fidelities_PD3 = []
fidelities_PD4 = []
range_PD = np.arange(0.001, 1, 0.001)
for p_d in range_PD:
    trace_1 = (2 * p_d + (1 - p_d) * MU * ETA * (1 - ETA))**2 + (1 - p_d)**2 * MU * ETA**2
    trace_2 = (2 * p_d + MU * ETA)**2 + MU * ETA**2
    trace_3 = (2 * p_d + (1 - 2 * p_d) * MU * ETA)**2 + (1 - 2 * p_d)**2 * MU * ETA**2
    trace_4 = (2 * p_d + (1 - p_d) * MU * ETA)**2 + MU * ETA**2

    fidelity_1 = ((p_d + 0.5 * (1 - p_d) * MU * ETA * (1 - ETA))**2 + (1 - p_d)**2 * MU * ETA**2) / trace_1
    fidelity_2 = ((p_d + 0.5 * MU * ETA)**2 + MU * ETA**2) / trace_2
    fidelity_3 = ((p_d + 0.5 * (1 - 2 * p_d) * MU * ETA)**2 + (1 - 2 * p_d)**2 * MU * ETA**2) / trace_3
    fidelity_4 = ((p_d + 0.5 * (1 - p_d) * MU * ETA)**2 + MU * ETA**2) / trace_4

    fidelities_PD1.append(fidelity_1)
    fidelities_PD2.append(fidelity_2)
    fidelities_PD3.append(fidelity_3)
    fidelities_PD4.append(fidelity_4)
print(fidelities_PD1)
print(fidelities_PD2)
print(fidelities_PD3)
print(fidelities_PD4)
```

```python
plt.plot(range_PD, fidelities_PD1)
plt.xlabel('Dark Count Probability')
plt.ylabel('Fidelity')
plt.title('Fidelity Plots for Varying Dark Count Probability, Case 1')
plt.show()
```

```python
plt.plot(range_PD, fidelities_PD2)
plt.xlabel('Dark Count Probability')
plt.ylabel('Fidelity')
plt.title('Fidelity Plots for Varying Dark Count Probability - Case 2')
plt.show()
```

```python
plt.plot(range_PD, fidelities_PD4)
plt.xlabel('Dark Count Probability')
plt.ylabel('Fidelity')
plt.title('Fidelity Plots for Varying Dark Count Probability, Case 4')
plt.show()
```

```python
# Plot fidelities based on varying dark count probability
plt.plot(range_PD, fidelities_PD1, label = 'Case 1')
plt.plot(range_PD, fidelities_PD2, label = 'Case 2')
plt.plot(range_PD, fidelities_PD3, label = 'Case 3')
plt.plot(range_PD, fidelities_PD4, label = 'Case 4')
plt.xlabel('Dark Count Probability')
plt.ylabel('Fidelity')
plt.title('Fidelity Plots for Varying Dark Count Probability')
plt.legend()
plt.savefig("fidelities_PD.png")
plt.show()
```
